```python
def error(u, y):
    u = u.flatten()
    y = y.flatten()
    return 0.5*np.sum((u - y)**2)
```

```python
# XOR function
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
n, p = X.shape
print(n)
print(p)
```

```python
# test error
error(np.zeros([3,1]), np.array([1,2,3]))
```

```python
# size of hidden layer
p1 = 4
# number of iterations
n_iter = 200
# learning rate
lr = 0.2
```

```python
# initialization
W1 = np.random.randn(p, p1)
W2 = np.random.randn(p1, 1)
for i in range(n_iter):
    # forward propagation (usually a for loop through layers)
    u1 = X@W1
    h1 = relu(u1)
    u2 = h1@W2
    
    # compute cost
    L = error(u2, y)
    print(L)

    # backward propogation for computing gradient
    g = u2 - y.reshape([n, 1])
    dW2 = (h1.T)@g
    dh = g@(W2.T)
    du1 = relu(dh)
    dW1 = (X.T)@du1

    # gradient descent
    W1 = W1 - lr * dW1
    W2 = W2 - lr * dW2
```

```python
model_nob = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(p,)),
  tf.keras.layers.Dense(p1, activation='relu'),
  tf.keras.layers.Dense(1, )
])
model_nob.summary()
```

```python
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=2e-1,
    decay_steps=20,
    decay_rate=0.8)
opt = tf.keras.optimizers.SGD(learning_rate=lr_schedule)

model_nob.compile(
    loss = 'mse',
    #optimizer= tf.keras.optimizers.SGD(learning_rate=0.2),
    optimizer = opt)
```

```python
history = model_nob.fit(
    X,
    y,
    batch_size=4,
    epochs=300,
    # We pass some validation for
    # monitoring validation loss and metrics
    # at the end of each epoch
    
)
```
