Let $A = USV^\top$ be the SVD of $A$.

Let $B=US$, then $A=BV^\top$

As an example, let $B=\begin{bmatrix}1&1\\0&1\\-1&1\\2&1\end{bmatrix}$, and $V=[v_1, v_2]$

So $A=BV^\top=\begin{bmatrix}1&1\\0&1\\-1&1\\2&1\end{bmatrix}\begin{bmatrix}v_1^\top\\v_2^\top\end{bmatrix}=\begin{bmatrix}v_1^\top+v_2^\top\\v_2^\top\\-v_1^\top +v_2^\top\\2v_1^\top +v_2^\top\end{bmatrix}$

This means **each row of A  $\in\text{span}(v_1,v_2)$**

## This is how SVD can give us best affine space fitting

### Question: Find the best line that fits (1,2,3), (1,-1,0), (2,3,-2), (1,1,1)

```python
# step 1: put these points as rows of a matrix
A = np.array([[1,2,3],
             [1,-1,0],
             [2,3,-2],
             [1,1,1]])
```

```python
# step 2: shift each point by (x_mean, y_mean, z_mean)
# this is the same as what was done in PCA
Ac = A - np.mean(A, axis=0)
Ac
```

```python
# step 3: take SVD of Ac
U,S,Vt=np.linalg.svd(Ac)
V=Vt.T
V
```

line is a 1-dim object, so we take the first col of V

all rows of Ac are approximately lying on span(v1) (a line going through origin)
