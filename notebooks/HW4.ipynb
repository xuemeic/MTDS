## \*5. Show that the spectral norm is a proper norm. 
This is to show that for any matrix $A, B\in\mathbb{R}^{m\times n}$, the spectral norm satisfies the following properties:
- $\|A\|_2\geq0$ and equality holds only when $A$ is the zero matrix.
- $\|\alpha A\|_2=|\alpha|\|A\|_2$ for any scalar $\alpha\in\mathbb{R}$
- $\|A+B\|_2\leq\|A\|_2+\|B\|_2$
