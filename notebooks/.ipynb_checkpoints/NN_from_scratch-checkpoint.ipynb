# Build a Simple Neural Network

```python
import numpy as np
```

## Define some functions

```python
def relu(x):
    # relu function
    return np.maximum(x, 0)
```

```python
# test relu
relu(np.array([[-2, 3],
              [5, -3]]))
```

```python
def cost_J(u, y):
    # MSE cost function: 1 / 2 * ||u-y||^2
    return 0.5*(np.sum((u - y)**2))
```

```python
# test cost
cost_J(np.array([1,2]), np.array([2,3]))
```
