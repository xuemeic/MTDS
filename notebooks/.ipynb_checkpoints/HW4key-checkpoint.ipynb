## 1. Modify the SVD notebook to perform background removal on the COLORED video.

import necessary packages

```python
import numpy as np
import matplotlib.pyplot as plt
from moviepy.editor import VideoFileClip 
```

### read the video

```python
video = VideoFileClip("../data/vid_background.avi")
```

```python
# shorter video, the first 50 seconds
vid = video.subclip(0, 50)
```

### turn this colored video to a 4D array

```python
vid_array = np.array(list(vid.iter_frames()))
vid_array.shape
```

### for each color channel, turn the 350 by 240 by 320 array into a matrix of 350 by (240*320)


```python
m = vid_array.shape[1]
n = vid_array.shape[2]
vid_R_mat = vid_array[...,0].reshape([-1, m*n]) # R channel
vid_G_mat = vid_array[...,1].reshape([-1, m*n]) # G channel
vid_B_mat = vid_array[...,2].reshape([-1, m*n]) # B channel
```

### take transpose

```python
vid_R_mat = vid_R_mat.T
vid_G_mat = vid_G_mat.T
vid_B_mat = vid_B_mat.T
vid_B_mat.shape
```

### for each color channel, take SVD

```python
U_r, S_r, Vt_r = np.linalg.svd(vid_R_mat, full_matrices=False) 
U_g, S_g, Vt_g = np.linalg.svd(vid_G_mat, full_matrices=False) 
U_b, S_b, Vt_b = np.linalg.svd(vid_B_mat, full_matrices=False) 
```

### best rank-1 approximation for each channel

```python
V_r = Vt_r.T; V_g = Vt_g.T; V_b = Vt_b.T
M1_r = S_r[0]*U_r[:,:1]@(V_r[:,:1].T)
M1_g = S_g[0]*U_g[:,:1]@(V_g[:,:1].T)
M1_b = S_b[0]*U_b[:,:1]@(V_b[:,:1].T)
```

### put three channels back into one array of (350, 240, 320, 3)

```python
M1_r.shape
```

```python
vid_bg = np.zeros(vid_array.shape)
vid_bg[...,0] = M1_r.T.reshape([-1,m,n])
vid_bg[...,1] = M1_g.T.reshape([-1,m,n])
vid_bg[...,2] = M1_b.T.reshape([-1,m,n])
```

### show a particular frame

```python
fidx = 1
mm = np.min(vid_array[fidx,...] - vid_bg[fidx,...].astype(int))
MM = np.max(vid_array[fidx,...] - vid_bg[fidx,...].astype(int))
moving_normalize = (vid_array[fidx,...] - vid_bg[fidx,...].astype(int) - mm)/(MM-mm)
fig, axs = plt.subplots(1, 3, figsize=(12,6)) 
axs[0].imshow(vid_array[fidx,...]);
axs[1].imshow(vid_bg[fidx,...].astype(int));
axs[2].imshow(moving_normalize);
```

## 4. Show that $\|A\|_2=\|A^\top\|_2$

Proof: Let $A=USV^\top$ be its full SVD, then $A^\top=(USV^\top)^\top=VS^\top U^\top$

$VS^\top U^\top$ is the SVD of $A^\top$ because 
- $V,U$ are orthogonal, and 
- $S^\top$ is still a diagonal matrix with nonnegative entries.

$\|A^\top\|_2$ = first entry of $S^\top$ = first entry of $S$ = $\|A\|_2$

## \*5. Show that the spectral norm is a proper norm. 
This is to show that for any matrix $A, B\in\mathbb{R}^{m\times n}$, the spectral norm satisfies the following properties:
- $\|A\|_2\geq0$ and equality holds only when $A$ is the zero matrix.
- $\|\alpha A\|_2=|\alpha|\|A\|_2$ for any scalar $\alpha\in\mathbb{R}$
- $\|A+B\|_2\leq\|A\|_2+\|B\|_2$

proof:

- positivity: $\|A\|_2=\sup_{\|x\|=1}\|Ax\|\geq0$

    if $\|A\|_2=\sup_{\|x\|=1}\|Ax\|=0$, then $\|Ax\|=0$ for every $x\neq0$, which forces $A$ to be the zero matrix.
    
- scalable: $\|\alpha A\|_2=\sup_{\|x\|=1}\|\alpha Ax\|=\sup_{\|x\|=1}|\alpha|\| Ax\|=|\alpha|\sup_{\|x\|=1}\| Ax\|=|\alpha|\|A\|_2$

- triangle inequality:

    $\|A+B\|_2=\sup_{\|x\|=1}\|(A+B)x\|\leq\sup_{\|x\|=1}(\|Ax\|+\|Bx\|)\leq (\sup_{\|x\|=1}\|Ax\|)+(\sup_{\|x\|=1}\|Bx\|)=\|A\|_2+\|B\|_2$
